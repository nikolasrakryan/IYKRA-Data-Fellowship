```python
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax)
        ax.set_title(feature+" Distribution",color='DarkRed')
        
    fig.tight_layout()  
    plt.show()
#draw_histograms(df,df.iloc[:,3:43],6,3)

try:
    draw_histograms(df,df.iloc[:,3:43],6,3)
except ValueError:
    pass  # do nothing!
```

###3.	Which team has the best attack? 

```python
df.columns
```

```python
#Selecting the relevant columns
attack_columns=['Team','attack_scored', 'attack_passes','attack_passes_through','attack_passes_long',
        'attack_passes_back','attack_crosses','attack_corners_taken', 
        'attack_shots','attack_shots_on_target','attack_goals_headed','attack_goals_penalty', 
        'attack_goals_box','attack_goals_outsidebox', 'attack_goals_counter','attack_goals_freekick']
df3=df[attack_columns]
df3
```

```python
df3["attack_mean_value"]=df3.mean(axis=1)
df3
```

```python
df_best_attack_team=df3[["Team","attack_mean_value"]].sort_values(by=["attack_mean_value"],ascending=False)
print(df_best_attack_team.head(5),"\n")

df_worst_attack_team=df3[["Team","attack_mean_value"]].sort_values(by=["attack_mean_value"])
print(df_worst_attack_team.head(5))
```

###4.	Which team has the best defence? 

```python
df.columns
```

```python
#Selecting the relevant columns
defence_columns=['Team','defence_saves',
       'defence_blocks', 'defence_interceptions', 'defence_tackles',
       'defence_tackles_last_man', 'defence_clearances',
       'defence_clearances_headed', 'defence_penalty_conceeded']
df4=df[defence_columns]
df4
```

```python
df4["defence_mean_value"]=df4.mean(axis=1)
df4
```

```python
df_best_defence_team=df4[["Team","defence_mean_value"]].sort_values(by=["defence_mean_value"],ascending=False)
print(df_best_defence_team.head(5),"\n")

df_worst_defence_team=df4[["Team","defence_mean_value"]].sort_values(by=["defence_mean_value"])
print(df_worst_defence_team.head(5))
```
